```python
cols = ['CustomerName','Email','Ph1Number','Pincode','OrderAmount','FinalAmount','ReviewText','ReturnReason','ResolutionStatus\r']
df.drop(cols, axis = 1, inplace = True)
```

```python
df['Discount'] = df['Discount'].str.replace('%', '', regex=False)
df['Discount'] = df['Discount'].fillna(0).astype(float).astype(int)
```

```python
df['Discount'].dtypes
```

```python
df['PurchaseDate'] = pd.to_datetime(df['PurchaseDate'], format='%d-%m-%Y', errors='coerce')
```

```python
df['PurchaseYear'] = df['PurchaseDate'].dt.year
```

```python
df['PurchaseMonth'] = df['PurchaseDate'].dt.month
```

```python
df['DeliveryDate'] = pd.to_datetime(df['DeliveryDate'], format='%d-%m-%Y', errors='coerce')
```

```python
df['DeliveryYear'] = df['DeliveryDate'].dt.year
```

```python
df['DeliveryMonth'] = df['DeliveryDate'].dt.month
```

```python
df.head(2)
```

```python
df['Discount'].fillna(df['Discount'].mean() , inplace = True)
```

```python
df['Discount'].isnull().sum()
```

```python
df['Rating'].fillna(df['Rating'].mean() , inplace = True)
```

```python
df['Rating'].isnull().sum()
```

```python
df['ComplaintRaised'] = df['ComplaintRaised'].fillna(df['ComplaintRaised'].mode()[0])
```

```python
df['PurchaseYear'] = df['PurchaseYear'].fillna(df['PurchaseYear'].mode()[0])
```

```python
Month_Year_Final_Amount = df.groupby(['PurchaseMonth','PurchaseYear'], as_index = False)['Final_Amount'].sum().sort_values(by = 'Final_Amount', ascending = False)
plt.figure(figsize = (18,6))
sns.barplot(data = Month_Year_Final_Amount, x = 'PurchaseMonth', y = 'Final_Amount', hue = 'PurchaseYear', palette = 'inferno')
plt.xticks(rotation=45)
plt.tight_layout()
plt.xlabel('PurchaseMonth')
plt.ylabel('Final_Amount')
plt.title('Month and Year wise Total sales')
plt.show()
```

```python
City_Gender_Final_Amount = df.groupby(['City','Gender'], as_index = False)['Final_Amount'].sum().sort_values(by = 'Final_Amount', ascending = False)
plt.figure(figsize = (18,6))
sns.barplot(data = City_Gender_Final_Amount, x = 'City', y = 'Final_Amount', hue = 'Gender', palette = 'ch:s=-.2,r=.6')
plt.xticks(rotation=45)
plt.tight_layout()
plt.xlabel('City')
plt.ylabel('Final_Amount')
plt.title('City and Gender wise Total sales')
plt.show()
```

```python
City_Age_Group_Final_Amount = df.groupby(['City','Age_Group'], as_index = False)['Final_Amount'].sum().sort_values(by = 'Final_Amount', ascending = False)
plt.figure(figsize = (18,6))
sns.barplot(data = City_Age_Group_Final_Amount, x = 'City', y = 'Final_Amount', hue = 'Age_Group', palette = 'flare')
plt.xticks(rotation=45)
plt.tight_layout()
plt.xlabel('City')
plt.ylabel('Final_Amount')
plt.title('City and Age_Group wise Total sales')
plt.show()
```
